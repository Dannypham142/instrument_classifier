```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv_path = 'logs/conv1d_history.csv'
df = pd.read_csv(csv_path)
```

```python
fig, ax = plt.subplots()
ax.set_title('Conv 1D')
ax.plot(df.accuracy, color='r', label='train')
ax.plot(df.val_accuracy, ls='--', color='r', label='test')
ax.legend(loc='upper left')
ax.set_xlabel('Epochs')
ax.set_ylabel('Accuracy')
ax.grid(True)
```
